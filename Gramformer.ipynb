# GRAMFORMER AI MODEL

```python
from gramformer import Gramformer
gf = Gramformer(models=1)
```

```python

def correct_text(passage):
    corrector = Gramformer(models=1)
    passage = passage.capitalize()
    sentences = passage.split('.') # split the passage into sentences based on period
    corrected_sentences = []
    for sentence in sentences:
        corrected_sentences.append(' '.join(list(corrector.correct(sentence)))) # convert set to list, join list items with space, and append to list
    return ' '.join(corrected_sentences) # concatenate the corrected sentences together with spaces between them
```

# TESTING

```python
passage = "My camera battery a dead. i plAYS football"
corrected_passage = correct_text(passage)
print(corrected_passage)
```

# GRADIO INTERFACE

```python
import gradio as gr

input_text = gr.inputs.Textbox(lines=10, label="Enter text to correct")
output_text = gr.outputs.Textbox(label="Corrected text")
examples = [
["I havve a job intervew tomoro."],
["The restrant was crowdy."],
["I need to by a new phone."],
["The airconditioner is not werking."],
["I forrgot to pack my luch."],
["I need to get a haircute."],
["I got lost and ended up at the libary."],
["I can't wait to see the florist's displaay."],
["I bought these shose last wek."],
["I hope the wheather stays nise for our picnik."]

]
gr.Interface(fn=correct_text, inputs=input_text, outputs=output_text, title="Text Corrector", examples=examples, description="Enter text to correct and click submit").launch(share=True)
```
